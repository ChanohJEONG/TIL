```python
import time

driver = wd.Chrome(executable_path='chromedriver.exe')
#driver = wd.PhantomJS(executable_path='phantomjs.exe')
driver.get('http://tour.interpark.com')

driver.find_element_by_id('box').send_keys('달럇')
driver.find_element_by_class_name('search-btn').click()
#######달럇 검색
driver.implicitly_wait(10)
driver.find_element_by_css_selector('.oTravelBox .moreBtn').click()
###### 여행지 더보기 클릭
time.sleep(1)
#driver.implicity_wait(5) # ajax 동작 코드까지는 못기다려줌 따라서 time.sleep()을 쓴것! 서버에서 값을 다 받았을때만 끝남

totalCnt = driver.find_element_by_id('totalAllCnt')
totalCnt = int(totalCnt.text)
totalPage = totalCnt // 10 + 1# 몫만 줌
if totalCnt % 10 == 0:
    totalPage -= 1 #게시글이 10으로 나눠질 경우 다음페이지로 인식하게 되어 있기 때문!

for i in range(5):
    time.sleep(1)
    product_list = driver.find_elements_by_css_selector('.panelZone .boxList .boxItem')
    for product in product_list:
        #product.find_elements_by_css_selector('proTit') 가져온 값중 제목만 뽑아오는 명령어
        print(product.text)
        print('========================================================')
    driver.find_element_by_css_selector('.pageNumBox .nextBtn').click()  
    
    
# driver.close()
```

```python
totalCnt = driver.find_element_by_id('totalAllCnt')
print(totalCnt.text)
totalCnt = int(totalCnt.text)
totalPage = totalCnt // 10 + 1# 몫만 줌
if totalCnt % 10 == 0:
    totalPage -= 1 #게시글이 10으로 나눠질 경우 다음페이지로 인식하게 되어 있기 때문!
```

```python
for page in range(1,totalPage + 1):
    driver.execute_script('searchModule.SetCategoryList(%d,'')' % page)
    time.sleep(1)
    product_list = driver.find_elements_by_css_selector('.panelZone .boxList .boxItem')
    for product in product_list:
        #product.find_elements_by_css_selector('proTit') 가져온 값중 제목만 뽑아오는 명령어
        print(product.text)
        print('========================================================')
    driver.find_element_by_css_selector('.pageNumBox .nextBtn').click()  
    
```
