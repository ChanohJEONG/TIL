## 데이터 연결하기(119쪽)

### 1. concat 메서드로 데이터 연결하기

```python
import pandas as pd

df1 = pd.read_csv('../data/concat_1.csv') 
df2 = pd.read_csv('../data/concat_2.csv') 
df3 = pd.read_csv('../data/concat_3.csv')
```

```python
row_concat = pd.concat([df1, df2, df3]) 
print(row_concat)
```

```python
# row_concat.iloc[3, ] # 단순히 3번째 row 값 출력
row_concat.loc[3, ]# index 값이 3인 row 모두 출력
```

### 4. 데이터프레임에 시리즈 연결하기

```python
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
new_row_series
```

```python
pd.concat([df1, new_row_series],axis=1,sort=False)# axis값이 0이면 해당 열에 값이 없기 때문에 결측치 값으로 나옴
```

```python
pd.concat([df1, new_row_series],axis=0)# axis값이 0이면 수직결합!
```

```python
new_row_series.name = 'E'#columm 명 지정
pd.merge(df1,new_row_series,left_index=True,right_index=True) # Join 연산은 컴퓨팅 비용이 비싸기 때문에 주로 concat 사용(단순결합용도)
```

### 6. 공통 열과 공통 인덱스만 연결하기

```python
df1.columns = ['A', 'B', 'C', 'D'] # column 명 재설정
df2.columns = ['E', 'F', 'G', 'H'] 
df3.columns = ['A', 'C', 'F', 'H']
print(df1)
print(type(df1))
```

```python
print(df2)
print(type(df2))
```

```python
print(df3)
print(type(df3))
```

```python
row_concat = pd.concat([df1, df2, df3]) 
print(row_concat)
```

```python
pd.concat([df1, df2, df3], join='inner')# 공통 열만 골라서 연결할때는 'inner' join!
```

```python
pd.concat([df1,df3], ignore_index=False, join='inner')
```

## merge 메서드 사용하기(129쪽)

```python
person = pd.read_csv('../data/survey_person.csv') 
site = pd.read_csv('../data/survey_site.csv') 
survey = pd.read_csv('../data/survey_survey.csv') 
visited = pd.read_csv('../data/survey_visited.csv')

person
```

```python
site
```

```python
visited
```

```python
survey
```

```python
visited_subset = visited.loc[[0, 2, 6], ]# visited 의 0,2,6 index 에 해당하는 값만 따로 추출 저장
visited_subset
```

```python
o2o_merge = site.merge(visited_subset, left_on='name', right_on='site') 
o2o_merge# site 의 name기준, 오른쪽 합병 데이터는 visited_subset의 리스트값중 site 기분
```

```python
m2o_merge = site.merge(visited, left_on='name', right_on='site') 
m2o_merge
```

```python
ps = person.merge(survey, left_on='ident', right_on='person') # left_on 과 right_on 에는 동일한 값이 있어야지만 정상적 merge 가능!
vs = visited.merge(survey, left_on='ident', right_on='taken')

ps# ident 값을 기준으로 합병! -danfort가 없어서 숫자가 적어보임!
```

```python
vs
```

```python
ps_vs = ps.merge(vs, left_on=['ident', 'taken', 'quant', 'reading'], right_on=['person', 'ident', 'quant', 'reading'])
```
